Reordering the columns of the DataFrame 'df' based on the specified list 're'

```python
re = ['Series_Title','Certificate', 'Runtime', 'Genre', 'IMDB_Rating', 'Meta_score', 'Director','No_of_Votes','Released_date','overview', 'crew', 'orig_lang', 'country', 'Gross', 'budget', 'revenue']
df = df [re]
```

```python
df.head()
```

Extracting numerical values from the 'Runtime' column and save it to the DataFrame 'df'

```python
df['Runtime'] =pd.to_numeric(df["Runtime"])
```

__pd.to_numeric():__ Converting the 'column name' column to numeric values

__Types of Error__<br>
__errors='coerce':__ return non-numeric values to NaN<br>
__errors='raise':__ raise an exception if a conversion error occurs.<br>
__errors='ignore':__ ignore the error, and pass it on to the next step in the pipeline (default).

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')
```

To make sure the convert is done

```python
df.info()
```

```python
df['Series_Title'] = df['Series_Title'].str.title()
```

Just confirming the change

```python
df.head()
```

To convert Released_date column the type from 'object' type to 'datetime64[ns]'

```python
df['Released_date'] = pd.to_datetime(df['Released_date'])
```
